## Exercice 2 : implémentation de plusieurs distances

L'objectif de cet exercice est d'écrire trois distances possibles afin de pouvoir programmer implémenter l'algorithme de $k$ plus proches voisin dans l'exercice suivant. Les trois distances choisies sont : la distance euclidienne, la distance de Manhattan et la distance de Tchebychev dont les formules sont données ci-dessous.

On considère deux données $d_1$ et $d_2$ de coordonnées respectives $(x_1, y_1)$ et $(x_2, y_2)$ dans un repère orthonormé. Voici les formules donnant les distances dans chaque cas :

- Distance **euclidienne** : $\text{distance}(d_1, d_2) = \sqrt{(x_2-x_1)^2 + (y_2-y_1)^2}$. 
- Distance **de Manhattan** : $\text{distance}(d_1, d_2) = |x_2-x_1| + |y_2-y_1|$.
- Distance **de Tchebychev** : $\text{distance}(d_1, d_2) = \text{max}\left(|x_2-x_1|, |y_2-y_1|\right)$.

### Distance euclidienne

La fonction `distance_euclidienne(d1, d2)` suivante renvoie la distance euclidienne entre les données `d1` et `d2` (tuples de deux éléments : abscisse et ordonnée).

````python
from math import sqrt # pour utiliser la fonction racine carrée = sqrt

def distance_euclidienne(d1, d2):
    x1 = d1['abs']  # abscisse de la donnée 1
    y1 = d1['ord']  # ordonnée de la donnée 1
    x2 = d2['abs']  # abscisse de la donnée 2
    y2 = d2['ord']  # ordonnée de la donnée 2
    return sqrt((x2-x1)**2 + (y2-y1)**2)  # formule donnant la distance euclidienne

assert distance_euclidienne(B, D) == 1  # la distance euclidienne entre les points (2,2) et (2,3) (B et D)
assert distance_euclidienne(A, B) == sqrt(2)
assert distance_euclidienne(D, D) == 0
assert distance_euclidienne(B, F) == sqrt(10)
````

On va travailler avec les données de l'exercice 1 dont on redonne la figure.

### Représentation des données

Comme dans l'exercice 2, chaque point est représente par un dictionnaire et on peut mémoriser dans une table `donnees` l'ensemble des points de la figure.

````python
A = {'nom': 'A', 'abs': 1, 'ord': 1, 'classe': 'Bleu'}
B = {'nom': 'B', 'abs': 2, 'ord': 2, 'classe': 'Bleu'}
C = {'nom': 'C', 'abs': 4, 'ord': 2, 'classe': 'Bleu'}
D = {'nom': 'D', 'abs': 2, 'ord': 3, 'classe': 'Rouge'}
E = {'nom': 'E', 'abs': 4, 'ord': 3, 'classe': 'Bleu'}
F = {'nom': 'F', 'abs': 3, 'ord': 5, 'classe': 'Rouge'}
G = {'nom': 'G', 'abs': 2, 'ord': 5, 'classe': 'Rouge'}
H = {'nom': 'H', 'abs': 6, 'ord': 4, 'classe': 'Bleu'}
I = {'nom': 'I', 'abs': 5, 'ord': 5, 'classe': 'Rouge'}

donnees = [A, B, C, D, E, F, G, H, I]
donnees
````

### Algorithme des kppv avec la distance euclidienne

On va maintenant implémenter l'algorithme des $k$ plus proches voisins (abrégé *kppv*) que l'on rappelle :

1. Créer une table `distances_voisins` contenant les éléments de la table `donnees` et leurs distances avec la donnée `cible`.  
2. Trier les données de la table `distances_voisins` selon la distance croissante avec la donnée `cible`
3. Renvoyer les `k` premiers éléments de cette table triée.

**Première étape : création de la table avec les distances** 

Il faut commencer par choisir comment représenter les éléments de la table `distances_voisins`. Ici, il a été décidé de reprendre les dictionnaires de la table `donnees` en leur ajoutant une clé correspondant à la distance avec la cible. Par exemple, le premier élément de `distances_voisins` sera un dictionnaire de la forme
```python
{'nom': 'A', 'abs': 1, 'ord': 1, 'classe': 'Bleu', 'distance' : dist}```

dans lequel `dist` est la distance à calculer entre le point `'A'` et la cible.

> On aurait aussi pu choisir d'utiliser un tableau ou même de ne conserver que les distances en travaillant par la suite avec leurs indices pour retrouver les plus proches voisins.

On peut construire la table `distances_voisins` de la façon suivante en prenant le point $M(1,3)$ comme point cible et la distance euclidienne comme distance.

````python
cible = {'nom': 'M', 'abs': 1, 'ord': 3}
 
distances_voisins = [0]*len(donnees)  # on construit un tableau de même taille que donnees
for i in range(len(donnees)):  # on parcourt les données pas leur indice i
    distances_voisins[i] = donnees[i]  # on les recopies dans la nouvelle table en position i
    distance = distance_euclidienne(donnees[i], cible)  # on calcule la distance entre la donnée et la cible
    distances_voisins[i]['distance'] = distance  # on ajoute la clé 'distance' qui vaut la valeur calculée au-dessus
````

On peut alors vérifier que la table `distances_voisins` contient bien les mêmes dictionnaires que `donnees` auxquels ont a ajouté la distance avec la cible avec une nouvelle clé.

````python
distances_voisins
````

**Deuxième étape : tri de la table**

Il faut désormais trier la table obtenue par ordre croissant des distances. Pour cela, on peut utiliser avantageusement la fonction `sorted`. On doit ici trier les dictionnaires selon leur clé `'distance'` donc il faut au préalable définir notre fonction de tri.


````python
def tri_distance(d):
    return d['distance']  # renvoie la valeur de la clé 'distance' d'un dictionnaire d
````

On peut alors trier notre table dans une nouvelle appelée `distances_voisins_triee`.

````python
distances_voisins_triee = sorted(distances_voisins, key=tri_distance)
distances_voisins_triee
````

On constate que les données ont bien été triées par ordre croissant des distances avec le point $M(1,3)$ : les points $D$, $B$, $A$, $G$, $F$, $E$, $C$, $I$ et $H$ dans cet ordre.

**Troisième étape : récupérer les $k$ plus proches voisins**

Pour récupérer les $k$ plus proches voisins il suffit de récupérer les $k$ premiers éléments de la table `distances_voisins_triee`. Avec une boucle `for` cela donnerait

````python
k = 3
k_plus_proches_voisins = [0]*k  # construction d'un tableau de taille k
for i in range(k): # dans lequel on écrit les k premières valeurs de la table distances_voisins_triee
    k_plus_proches_voisins[i] = distances_voisins_triee[i]  
print(k_plus_proches_voisins)
````

On peut également faire cela de manière plus rapide en utilisant un "slice".

````python
k = 3
k_plus_proches_voisins = distances_voisins_triee[0:k]  # récupère les k premiers éléments de distances_voisins_triee
print(k_plus_proches_voisins)
````

On constate, ce que l'on peut facilement vérifier sur le graphique, que : 
- les trois points les plus proches de $M$ sont les points $D(2,3)$, $B(2,2)$ et $A(1,1)$ ;
- la classe majoritaire parmi ces voisins est `'Bleu'` (2 contre 1) donc on peut prédire que le point cible $M$ est de classe "Bleue".

**BILAN**

On peut alors réunir ces trois parties pour créer une fonction `kppv(donnees, cible, k)` qui prend en paramètres la table de données `donnees`, la donnée `cible` et le nombre `k` de voisins choisis. On obtient alors ceci :

````python
def tri_distance(d):  # clé du tri (selon la clé 'distance')
    return d['distance']

def kppv(donnees, cible, k):
    # étape 1 : création de la table avec les distances
    distances_voisins = [0]*len(donnees)
    for i in range(len(donnees)):
        distances_voisins[i] = donnees[i]
        distance = distance_euclidienne(donnees[i], cible)
        distances_voisins[i]['distance'] = distance
    # étape 2 : tri par distance croissante
    distances_voisins_triee = sorted(distances_voisins, key=tri_distance)
    # étape 3 : récupération de k plus proches voisins que l'on renvoie
    k_plus_proches_voisins = distances_voisins_triee[0:k]
    return k_plus_proches_voisins
````

On peut vérifier que l'on obtient bien ce qui a été trouvé dans l'exercice 1. Par exemple, affichons les 3 plus proches voisins du point $N$.

````python
donnees = [A, B, C, D, E, F, G, H, I]
cible = {'nom': 'M', 'abs': 1, 'ord': 3}
k = 3
kppv(donnees, cible, k)
````
