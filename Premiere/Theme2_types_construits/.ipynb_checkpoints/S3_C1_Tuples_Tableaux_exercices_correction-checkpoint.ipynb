## Exercice 1 :
On considère la variable ``repertoire`` suivante.

```python
repertoire = [('Michel', '0210101010'), ('Marjorie', '0211111111'), ('Perrine', '0212121212')]
```

1. Quel est le type de la variable ``repertoire``?
2. Quel est le type de la variable ``repertoire[0]``?
3. Qu'affiche l'instruction ``repertoire[1][1]`` ?
4. Quelle instruction permet d'accéder au numéro de téléphone de Michel ?
5. Quel est le type de la variable ``repertoire[2][0]``?

**Réponses:**
1. tableau (de tuples)
2. tuple (de 2 chaines de caractères)
3. '0211111111'
4. `repertoire[0][1]` (deuxième élément du premier élément du tableau)
5. chaine de caractères ('Perrine')

## Exercice 7 :
Ecrire une fonction ``ajout(v, T)`` qui crée un nouveau tableau contenant d'abord tous les éléménts du tableau ``T`` puis ``v``.

```python
def ajout(v, T):
    tab = [0]*(len(T) + 1)
    for i in range(len(T)):
        tab[i] = T[i]
    tab[len(T)] = v
    return tab

# Variante : on remplit le tableau initial de valeurs v et non pas de 0:

def ajoutbis(v, T):
    tab = [v]*(len(T) + 1)
    for i in range(len(T)):
        tab[i] = T[i]
    return tab

# ESSAI
t = [12, 15, 2, 1]
v = 7
nouv_tab = ajout(v, t)
print(nouv_tab)
nouv_tab = ajoutbis(v,t)
print(nouv_tab)

#---------- AUTRE METHODE : avec append ---------------
def ajoutter(v, T):
    tab = []
    for i in range(len(T)):
        tab.append(T[i])
    tab.append(v)
    return tab

# ESSAI
t = [12, 15, 2, 1]
v = 7
nouv_tab = ajoutter(v, t)
print(nouv_tab)
```

## Exercice 8 :
Ecrire une fonction ``concatenation(t1, t2)`` qui crée un nouveau tableau contenant, dans l'ordre, tous les éléments du tableau ``t1`` puis tous les éléments du tableau ``t2``.

```python
def concatenation(t1, t2):
    tab = [0]*(len(t1) + len(t2))
    for i in range(len(t1)):
        tab[i] = t1[i]
    for i in range(len(t2)):
        tab[len(t1) + i] = t2[i]
    return tab

# ESSAI
t1 = [1, 2, 3, 4]
t2 = [10, 9, 8, 7]
nouv_tab = concatenation(t1, t2)
nouv_tab
```
